```python
import requests
import pandas as pd
import geopandas as gpd
import shapely
```

```python
resp=requests.get(url="https://opengis.regina.ca/arcgis/rest/services/CGISViewer/TreeWebApp/MapServer/0/query?outFields=*",
                     params=dict(
                         f='json',
                         returnGeometry='true',
                         spatialRel='esriSpatialRelIntersects',
                         where='(1=1) AND (1=1)',
                         orderByFields='GLOBALID ASC',
                         outSR=4326,
                         resultOffset=0,
                         resultRecordCount=1
                     ))
resp.json()
```

```python
results = []
nbr_results = 1
while nbr_results>0:
    resp=requests.get(url="https://opengis.regina.ca/arcgis/rest/services/CGISViewer/TreeWebApp/MapServer/0/query?outFields=*",
                     params=dict(
                         f='json',
                         returnGeometry='true',
                         spatialRel='esriSpatialRelIntersects',
                         where='(1=1) AND (1=1)',
                         orderByFields='GLOBALID ASC',
                         outSR=4326,
                         resultOffset=len(results),
                         resultRecordCount=10000,
                         geometryType='esriGeometryPoint'
                     ))
    nbr_results=len(resp.json()['features'])
    results.extend(resp.json()['features'])
    print(len(results))
```

```python
len(results)
```

```python
data = pd.DataFrame([{**doc['attributes'],"geometry":doc.get('geometry')} for doc in results])
```

```python
data=data[data.geometry.isna()==False]
```
